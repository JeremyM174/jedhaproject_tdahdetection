## 2. generation du df 

```python
# === Données ===
datatrain = mat['train'][0]  # ou ['val'][0] ou ['test'][0] selon ton split

# === Stockage final ===
dataset = []

for i in range(17077):
    img_name = datatrain[i]['filename'][0]
    image_size_ncol = int(datatrain['image_size'][i]['n_col'][0][0][0][0])
    image_size_nrow = int(datatrain['image_size'][i]['n_row'][0][0][0][0])
    
    persons = datatrain[i]['person']
    for person_id, person in enumerate(persons):
        # --- Bounding box ---
        bbox_x1 = int(person['body_bbox'][0][0][0])
        bbox_y1 = int(person['body_bbox'][0][0][1])
        bbox_x2 = int(person['body_bbox'][0][0][2])
        bbox_y2 = int(person['body_bbox'][0][0][3])   # [x1, y1, x2, y2]
        
        
        # --- Catégories d’émotion (multilabel) ---
        categories_array = person['annotations_categories'][0][0][0]['categories'][0]
        categories = [str(c[0]) for c in categories_array]
        
        # --- Émotions continues ---
        annotations = person['annotations_continuous'][0]
        arousal = float(annotations['arousal'][0][0][0][0])
        valence = float(annotations['valence'][0][0][0][0])
        dominance = float(annotations['dominance'][0][0][0][0])
        
        dataset.append({
            'image': img_name,
            'height': image_size_nrow,
            'width': image_size_ncol,
            'person_id': person_id,
            'body_bbox_x1': bbox_x1,
            'body_bbox_y1': bbox_y1,
            'body_bbox_x2': bbox_x2,
            'body_bbox_y2': bbox_y2,
            'emotions': categories,
            'arousal': arousal,
            'valence': valence,
            'dominance': dominance
        })
    
df = pd.DataFrame(dataset)
df
```

## 2. création de colonnes emotions 

```python
# 1. Identifier toutes les émotions uniques
# Utilise .explode() pour transformer la liste en lignes individuelles, puis .unique()
toutes_emotions_uniques = df_copy1['emotions'].explode().unique()
# Retire les valeurs NaN si la liste 'Emotions' peut être vide ou contenir des NaN
toutes_emotions_uniques = [emo for emo in toutes_emotions_uniques if pd.notna(emo)]

print(f"\nToutes les émotions uniques trouvées : {toutes_emotions_uniques}")

# 2. Créer les nouvelles colonnes binaires
# Pour chaque émotion unique, on va créer une colonne et la remplir avec 0 initialement
for emotion in toutes_emotions_uniques:
    # Utilise une list comprehension pour vérifier la présence de l'émotion dans chaque liste
    df_copy1[emotion] = [1 if emotion in row_emotions else 0 for row_emotions in df_copy1['emotions']]

print("\nDataFrame après encodage One-Hot/Multi-Hot :")
print(df_copy1)
```
